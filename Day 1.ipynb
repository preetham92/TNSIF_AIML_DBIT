```python
def atm_machine():
    password = '1928'
    balance = 2000
    attempts = 3
    
    while attempts > 0:
        pin = input("Enter your password: ")
        if pin == password:
            print("\n Login Successful!\n")
            
            while True:
                print("======Menu=======")
                print("1. Withdraw")
                print("2. Deposit")
                print("3. Check Balance")
                print("4. Logout")
                
                option = input("Choose an option (1-4): ")
                
                if option == '1':
                    amount = float(input("Enter amount to withdraw: "))
                    if amount > balance:
                        print("Insufficient balnce. ")
                    else:
                        balance -= amount
                        print(f'Withdrawal successful. Remaining balance: Rs {balance}')
                        
                elif option == '2':
                    amount = float(input("Enter amount to deposit: "))
                    if amount <= 0:
                        print("Amount should be greater than 0....")
                    else:
                        balance += amount
                        print(f"Deposit successful. New Balance: Rs {balance}")
                
                elif option == '3':
                    print(f"Current Balance: Rs {balance}")
                
                elif option == '4':
                    print("Thank you! You have been logged out.")
                    
                    return 
                else:
                    print("Invalid choice Please select 1-4.")
        else:
            attempts -= 1
            print(f"Wrong password. attempts left: {attempts}")
    
    print("Your account hase been blocked! Contact customer care.")
    

atm_machine()
                    
                        
                
```
